```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:2].values
Y = dataset.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_10 = RandomForestRegressor(n_estimators = 10, random_state = 0) # N_estimator is number of decision tree we want
regressor_100 = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor_300 = RandomForestRegressor(n_estimators = 300, random_state = 0)

# made seperate regressor for each number of tree 10,100,300

regressor_10.fit(X,Y)
regressor_100.fit(X,Y)
regressor_300.fit(X,Y)
```

```python
y_pred_10 = regressor_10.predict(np.array(6.5).reshape(1,-1))
y_pred_10
```

```python
y_pred_100 = regressor_100.predict(np.array(6.5).reshape(1,-1))
y_pred_100
```

```python
y_pred_300 = regressor_300.predict(np.array(6.5).reshape(1,-1))
y_pred_300
```

```python
# So we see we getting closer and closer for everytime we increase the number of decision tree
# Lets see in graph
```

```python
X_grid = np.arange(min(X),max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y, color = 'red')
plt.plot(X_grid,regressor_10.predict(X_grid), color = 'blue')
plt.title('Random Forest (10 Trees)')
plt.xlabel('Levels')
plt.ylabel('Salaries')
plt.show()
```

```python
plt.scatter(X,Y, color = 'red')
plt.plot(X_grid,regressor_100.predict(X_grid), color = 'blue')
plt.title('Random Forest (100 Trees)')
plt.xlabel('Levels')
plt.ylabel('Salaries')
plt.show()
```

```python
plt.scatter(X,Y, color = 'red')
plt.plot(X_grid,regressor_300.predict(X_grid), color = 'blue')
plt.title('Random Forest (300 Trees)')
plt.xlabel('Levels')
plt.ylabel('Salaries')
plt.show()
```

```python
# As we see number of stairs doesnot grow as fast as we increasing number of tree it is because more number of tree
# helping to get the average more coverging so number of steps increase but not so much but it give much closer values
```
